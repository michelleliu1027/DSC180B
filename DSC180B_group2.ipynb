```python
battery_event = pd.read_csv('sample_batt_acdc_events.csv')
battery_event = battery_event[battery_event.columns[1:]]
```

```python
battery_info = pd.read_csv('sample_batt_info.csv')
battery_info = battery_info[battery_info.columns[1:]]
```

```python
df3 = pd.read_csv('sample_devuse_4known_device.csv')
df3 = df3[df3.columns[1:]]
df4 = pd.read_csv('sample_devuse_4known_device2.csv')
df4 = df4[df4.columns[1:]]
device_use = df4.append(df3)
```

```python
## First Guessing: number of device would affect the duration of battery (a way for evaluating the performance of battery)
device_cnt = device_use.groupby('dt')['name'].count()
device_cnt = device_cnt[set(battery_event['dt']).intersection(device_use['dt'])]
battery_w_device = battery_event.set_index('dt').loc[set(battery_event['dt']).intersection(device_use['dt'])].reset_index()
data1 = battery_w_device.groupby('dt')['duration'].mean()
data2 = device_cnt.sort_index()


fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('date')
ax1.set_ylabel('avg duration', color=color)
ax1.plot(data1.iloc[1:].index, data1.iloc[1:], color=color) 
## need to filter out 1/1/2020, which is the new year, so special day would remain outliers, affect our analysis
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('device counts', color=color)  # we already handled the x-label with ax1
ax2.plot(data1.iloc[1:].index, data2.iloc[1:], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

## From the graph, we could notice that it seems like there's no strong relationship between the number of devices
## and the duration of battery. But we could notice that : with the increment of 
## number of devices, the duration of battery would be lower
## so switch to correlation analysis
```
