```python
import pandas as pd
import gzip
```

```python
# read the sample data
with gzip.open('batt_acdc_events.csv000.gz','rt') as f:
    count = 0
    for line in f:
        if count == 0:
            toreplace = line[7] 
            print(toreplace)
            count+=1
        else:
            break
            
df = pd.read_csv('batt_acdc_events.csv000.gz', compression='gzip', header=0, sep=toreplace,nrows = 100000)
df.to_csv('sample_batt_acdc_events.csv')
```
